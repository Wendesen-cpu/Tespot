```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
wende = cv2.imread("wende.jpeg",0)
```

```python
plt.imshow(wende, cmap='gray')
```

```python
def flip(img, dir):
  row,col = img.shape
  transposed_img = np.zeros((row,col),dtype=img.dtype)
  for i in range(row):
    for j in range(col):
      if(dir=="vertical"):
         transposed_img[i,j]=img[i,col-j-1]
      else:
        transposed_img[i,j]=img[row-i-1,j]
  return transposed_img
```

```python
wende_flipped_vertical = flip(wende,"vertical")
wende_flipped_horizontal = flip(wende,"horizonta")
```

```python
plt.imshow(wende_flipped_vertical, cmap='gray',vmin=0, vmax=255)
```

```python
plt.imshow(wende_flipped_horizontal, cmap='gray')
```

```python
def transpose(img):
  row,col = img.shape
  transposed_img = np.zeros((col,row),dtype=img.dtype)
  for i in range(row):
    for j in range(col):
         transposed_img[j,i]=img[i,j] 
  return transposed_img
```

```python
wende_transposed = transpose(wende)
plt.imshow(transpose(wende), cmap="gray")
```

```python
def crop_img(img,h_init,row,v_init,col):
  x,y=img.shape
  if(h_init>0 and row < x+row and v_init >0 and v_init + col < y):
    cropped_img = img[h_init:h_init+row,v_init: v_init+col]
  return cropped_img

```

```python
wende_cropped = crop_img(wende,150,1000,350,1200)
plt.imshow(wende_cropped, cmap="gray")
```

```python
def conv_2D(img,kernel,output_type):
  img_row,img_col = img.shape
  kr_row,kr_col = kernel.shape
  c = np.zeros((img_row+kr_row-1,img_col+kr_col-1), dtype=np.float)
  img = img.astype(np.float)
  for i in range(img_row):
    for h in range(kr_row):
      for j in range(kr_row):
        for k in range(kr_col):
          c[i+h][j+k]= c[i+h][j+k]+img[i,j]*kernel[h,k]
  c = c.astype(output_type)
  return c
```

```python
test = np.array([[-0.25,0.333,0.25],[-0.5,0.333,0.5],[-0.25,0.333,0.25]])
```

```python
my_2d = conv_2D(wende,test,np.uint8)
```

```python
def img_conv(img,kernel):
  img_row,img_col = img.shape
  ker_row,ker_col = kernel.shape

  m = int(ker_row/2)
  n = int(ker_col/2)

  img_conv = np.zeros(img.shape)
  for i in range(m):
    for j in range(n):
      img_conv[i,j]=img[i,j]

  for i in range(img_row-m,img_row):
    for j in range(img_col-n,img_col):
      img_conv[i,j]=img[i,j]


    

  for i in range(m,img_row-m):
    for j in range(n,img_col-n):
      sum = 0
      for h in range(ker_row):
        for k in range(ker_col):
          sum = sum + kernel[h,k]*img[i-m+h,j-n+k]
      img_conv[i,j] = sum
  return img_conv
```

```python
convolved_wende = img_conv(wende,kernel_2)
```
